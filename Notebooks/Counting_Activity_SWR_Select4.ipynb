## Produce two plots based on each optimal single-site structure
* Energies
* The counting sites at different potentials

$ V_{anode} = 0.31 V $

All Pt-based perfect. Copper is the cheapest, so I'll show that

```python
potential = 0.31
metal_A, metal_B = "Pt", "Cu"

SWR_surface = initialize_swim_surface([metal_A], [metal_B], dim_x, dim_y)
SWR_surface = precompute_binding_energies_SPEED2(SWR_surface, dim_x, dim_y, models)

active, inactive, blocked = counting_activity(SWR_surface, potential)
active_percent = active/(dim_x*dim_y)
print(f"{active_percent:.3%}")
power_per_site, open_circuit_current = power_puff(potential, active_percent)
print(f"Power per site: {power_per_site:.2f}")
print(f"Fuel cell potential: {open_circuit_current:.2f}")

filename = "Best_At_Potential_SS/0_31/Energies"
reward_type = "None"

deltaEdeltaE_plot_potential(filename, SWR_surface, potential, pure_metal_info, reward_type = None, show_plot=True)
```

```python
# Calculate 
potential_range, active_list, inactive_list, blocked_list = counting_activity_scan(SWR_surface, 0.1, 0.35, 100)

split = list(SWR_surface["stochiometry"].values())

# Plot and save
filename = "../figures/Best_At_Potential_SS/0_31/Potential_range.png"
counting_activity_plot(potential_range, active_list, inactive_list, blocked_list, potential, split, filename)
```
